```python
# 导入核心库
import pandas as pd
import numpy as np

# 设置文件路径（根据实际位置修改）
file_path = "./PTSD.xlsx"  # 如果文件在子文件夹中，例如："./datasets/PTSD.xlsx"

# 读取Excel文件
try:
    df = pd.read_excel(
        file_path,
        engine='openpyxl',  # 必须指定engine处理xlsx格式
        sheet_name=0        # 读取第一个sheet，可按需修改名称或序号
    )
    print("数据集加载成功！共读取 {} 行 {} 列".format(*df.shape))
    
except Exception as e:
    print("文件读取失败，请检查路径和文件格式:", e)

# 数据预览
if 'df' in locals():
    # 显示前3行数据
    print("\n数据预览：")
    display(df.head(3))
    
    # 显示基础信息
    print("\n数据结构：")
    df.info()
    
    # 基础统计
    print("\n数值列统计：")
    display(df.describe())
    
    # 缺失值检查
    print("\n缺失值统计：")
    display(df.isnull().sum())
    
    # 删除完全重复行
    initial_count = len(df)
    df.drop_duplicates(inplace=True)
    print(f"\n删除重复行：{initial_count} → {len(df)} 行")

# 可选：保存预处理结果（按需取消注释）
# df.to_excel("processed_data.xlsx", index=False)
```

```python
# 初步的筛选,直接移除没意义的列
# 删除指定列（在已加载df的前提下）
try:
    # 建议先确认列名是否完全匹配
    print("原始列名：", df.columns.tolist())
    
    # 执行删除操作
    df.drop(columns=['序号', '籍贯', '民族'], inplace=True, errors='ignore')
    
    print("\n删除后列名：", df.columns.tolist())
    print(f"剩余列数：{len(df.columns)}")
    
except Exception as e:
    print("列删除时发生错误:", e)

# 查看删除后数据（可选）
if not df.empty:
    display(df.head(5))
```

```python
# === 数据类型诊断与转换 ===
# 备份原始数据
df_orig = df.copy()

# 1. 查看当前类型概况
print("=== 数据类型概览 ===")
print(df.dtypes)


# 2. 自动识别变量类型
numeric_cols = df.select_dtypes(include=['number']).columns.tolist()
categorical_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()

# 3. 手动修正特殊类型（根据业务逻辑）
type_adjustments = {
    'nominal': [
        '婚姻状况', '性别','ASD性质','遗传史',
        '近期是否出现意外事件', '是否有家庭成员或亲近朋友意外',
        '意外事件中是否目睹有人重伤', '意外事件及救援过程中，您是否接触或见过尸体',
        '您是否曾因为灾难现场的景象', '既往是否有精神病史',
        '近一月内是否使用药物', '是否吸烟（包括以前）',
        '吸烟状态', '是否饮酒'
    ],
    'ordinal': {
        # 修复1：确保类别顺序与原始数据类型一致
        '文化程度': [3, 2, 1],          # 使用int类型匹配原始数据
        '家庭人均月收入': [1, 2, 3, 4]   # 使用int类型匹配原始数据
    },
}

# 4. 执行类型转换（修复类型匹配问题）
for col, order in type_adjustments['ordinal'].items():
    # 保持原始int64类型直接转换
    df[col] = pd.Categorical(
        df[col],
        categories=order,
        ordered=True,
    )
for col in type_adjustments['nominal']:
    if col in df.columns:
        # 将int64直接转为无序分类
        df[col] = pd.Categorical(df[col], categories=df[col].unique(), ordered=False)

#数据概览
print(df.dtypes)

# 新增验证步骤（在转换后立即执行）
print("\n=== 有序分类验证 ===")
for col in ['文化程度', '家庭人均月收入']:
    if col in df.columns:
        print(f"{col} 类别顺序: {df[col].cat.categories.tolist()}")
        print(f"是否有序: {df[col].cat.ordered}")
        print(f"数据类型: {df[col].dtype}\n")


```

```python
#基于相关性的进一步筛选,移除低相关特征以及高度自相关特征,这一阶段的筛选仍旧基于习惯的考虑,而不是全盘考虑
# 计算相关系数及统计指标
from scipy import stats
import pandas as pd

# 筛选有效数据
analysis_df = df[['年龄', '身高', '体重', 'BMI', 'ASD总分', '心理韧性', 'PCL总分']].dropna()

# 初始化结果存储
results = []

# 遍历每个预测变量
for var in analysis_df.columns[:-1]:  # 排除最后的PCL总分
    # 计算Pearson相关系数
    r, p = stats.pearsonr(analysis_df[var], analysis_df['PCL总分'])
    
    # 计算t统计量
    n = len(analysis_df)
    t = r * ((n - 2) ** 0.5) / ((1 - r**2) ** 0.5)
    
    # 确定显著性标记
    sig = ''
    if p < 0.001: sig = '***'
    elif p < 0.01: sig = '**'
    elif p < 0.05: sig = '*'
    
    results.append({
        '变量': var,
        '相关系数类型': 'Pearson r',
        '检验类型': 't检验',
        '统计量': round(t, 3),
        '相关系数': f"{round(r, 3)}{sig}",
        'P值': round(p, 4),
        '样本量': n
    })

# 创建结果表格
result_table = pd.DataFrame(results).set_index('变量')

# 格式化输出
display(result_table.style.format({
    '统计量': '{:.3f}',
    'P值': '{:.4f}'
}).set_caption("变量与PCL总分的相关系数分析表 (***p<0.001, **p<0.01, *p<0.05)"))


# 修改后的点二列相关分析
from scipy.stats import pointbiserialr

# 1. ASD总分与ASD性质相关性
asd_results = []
if 'ASD性质' in df.columns and 'ASD总分' in df.columns:
    # 清理数据
    valid_data = df[['ASD性质', 'ASD总分']].dropna()
    
    # 计算点二列相关
    r, p = pointbiserialr(valid_data['ASD性质'], valid_data['ASD总分'])
    
    # 记录结果
    asd_results.append({
        '变量对': 'ASD总分 vs ASD性质',
        '相关系数(r)': round(r, 3),
        '检验类型': '点二列相关',
        '统计量': f"r={r:.3f}",
        'P值': f"{p:.4f}{'***' if p < 0.001 else '**' if p < 0.01 else '*' if p < 0.05 else ''}",
        '样本量': len(valid_data)
    })

# 2. 分类变量与PCL总分相关性
cat_vars = ['遗传史', '性别', 'ASD性质']
pcl_results = []

for var in cat_vars:
    if var in df.columns:
        valid_data = df[[var, 'PCL总分']].dropna()
        
        # 计算点二列相关
        r, p = pointbiserialr(valid_data[var], valid_data['PCL总分'])
        
        pcl_results.append({
            '变量': var,
            '相关系数(r)': round(r, 3),
            '检验类型': '点二列相关',
            '统计量': f"r={r:.3f}",
            'P值': f"{p:.4f}{'***' if p < 0.001 else '**' if p < 0.01 else '*' if p < 0.05 else ''}",
            '样本量': len(valid_data)
        })

# 结果展示
print("=== ASD总分与性质相关性 ===")
display(pd.DataFrame(asd_results))

print("\n=== 分类变量与PCL总分相关性 ===")
display(pd.DataFrame(pcl_results))




```

根据前面进行的相关性检验,我们决定移除身高与体重变量,因为它们与PCL总分的相关系数均小于0.1,且P值均大于0.05,因此我们认为这两个变量与PCL总分没有显著的相关性,可以移除这两个变量。保留了BMI变量,因为其他相关研究中认为BMI与PCL总分存在相关性,可能是因为简单的线性模型无法拟合复杂的关系导致相关性不强。同时,由于自相关性(ASD总分就是子项分数的和),我们移除了ASD各个指标中除了总分以外的所有指标.经过初步的筛选,我们保留了20个列

```python
# 移除指定列
columns_to_drop = ['身高', '体重','吸烟状态','ASD总分','遗传史'] 
# 添加所有以ASD开头但不是总分的列
asd_cols = [col for col in df.columns if col.startswith('ASD') and col != 'ASD总分']
columns_to_drop += asd_cols

# 执行删除（安全模式，忽略不存在的列）
df.drop(columns=columns_to_drop, inplace=True, errors='ignore')

# 验证删除结果
print("剩余列名：", df.columns.tolist())
print("剩余列数：", len(df.columns))

# 创建完整列名映射字典
column_mapping = {
    '年龄': 'age',
    '性别': 'gender',
    '文化程度': 'education',
    '婚姻状况': 'marital_status',
    '家庭人均月收入': 'income',
    '近期是否出现意外事件': 'recent_accident',
    '是否有家庭成员或亲近朋友意外': 'family_accident',
    '意外事件中是否目睹有人重伤': 'witnessed_injury',
    '意外事件及救援过程中，您是否接触或见过尸体': 'witnessed_corpse',
    '您是否曾因为灾难现场的景象': 'traumatic_scene',
    '既往是否有精神病史': 'psychiatric_history',
    '近一月内是否使用药物': 'medication',
    '是否吸烟（包括以前）': 'smoking',
    '是否饮酒': 'alcohol',
    'PCL总分': 'pcl_scores',
    '心理韧性': 'resilience',
    'BMI': 'bmi',
}

# 执行列名转换
df.rename(columns=column_mapping, inplace=True)

# 验证转换结果
print("当前列名:", df.columns.tolist())
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from pandas.api.types import is_numeric_dtype
from scipy.stats import chi2_contingency

# 假设您的19列数据框 'df' 已经加载
# 例如:
# df = pd.read_csv('your_data.csv')

# --- 1. 定义计算关联度的函数 (与之前相同) ---
def cramers_v(x, y):
    """ 计算两个类别变量之间的Cramér's V关联度 """
    confusion_matrix = pd.crosstab(x, y)
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))
    rcorr = r - ((r-1)**2)/(n-1)
    kcorr = k - ((k-1)**2)/(n-1)
    # 修正分母为0的情况
    if min((kcorr-1), (rcorr-1)) == 0:
        return 0
    else:
        return np.sqrt(phi2corr / min((kcorr-1), (rcorr-1)))

def correlation_ratio(categorical, numerical):
    """ 计算类别变量和数值变量之间的Correlation Ratio (Eta) """
    f_cat, _ = pd.factorize(categorical)
    cat_num = np.asarray(f_cat)
    y_avg_array = np.zeros(cat_num.max() + 1)
    n_array = np.zeros(cat_num.max() + 1)
    for i in range(0, cat_num.max() + 1):
        cat_measures = numerical[cat_num == i]
        n_array[i] = len(cat_measures)
        y_avg_array[i] = np.average(cat_measures)
    y_total_avg = np.sum(np.multiply(y_avg_array, n_array)) / np.sum(n_array)
    numerator = np.sum(np.multiply(n_array, np.power(np.subtract(y_avg_array, y_total_avg), 2)))
    denominator = np.sum(np.power(np.subtract(numerical, y_total_avg), 2))
    if denominator == 0 or numerator == 0:
        eta = 0.0
    else:
        eta = np.sqrt(numerator / denominator)
    return eta

# --- 2. 修改 get_association_matrix 函数以返回绝对值 ---
def get_association_matrix(df):
    """ 
    计算包含所有变量类型的完整关联矩阵。
    所有关联度都取绝对值，范围在 [0, 1] 之间。
    """
    cols = df.columns
    n = len(cols)
    assoc_matrix = pd.DataFrame(np.zeros((n, n)), index=cols, columns=cols)

    for i in range(n):
        for j in range(i, n): # 优化循环，因为矩阵是对称的
            if i == j:
                assoc_matrix.iloc[i, j] = 1.0
                continue

            col1 = df[cols[i]]
            col2 = df[cols[j]]
            
            is_num1 = is_numeric_dtype(col1)
            is_num2 = is_numeric_dtype(col2)
            
            val = 0.0
            if is_num1 and is_num2:
                # 数值 vs 数值: 取皮尔逊相关系数的绝对值
                val = abs(col1.corr(col2))
            elif not is_num1 and not is_num2:
                # 类别 vs 类别
                val = cramers_v(col1, col2)
            elif is_num1 and not is_num2:
                # 数值 vs 类别
                val = correlation_ratio(col2, col1)
            else:
                # 类别 vs 数值
                val = correlation_ratio(col1, col2)
            
            assoc_matrix.iloc[i, j] = val
            assoc_matrix.iloc[j, i] = val # 填充对称位置
            
    return assoc_matrix

# 计算完整的、绝对值化的关联矩阵
association_matrix = get_association_matrix(df)


# --- 3. 使用新的关联矩阵和绘图逻辑进行可视化 ---

# 解决19x19问题：通过 `vars` 参数强制使用所有列
g = sns.PairGrid(df, vars=df.columns.tolist(), diag_sharey=False)
g.fig.suptitle("Absolute Association Matrix", y=1.02, fontsize=16)

# 右上部分: 修改为使用顺序调色板的热力图
def plot_assoc_heatmap_absolute(x, y, **kwargs):
    ax = plt.gca()
    ax.set_axis_off()
    
    value = association_matrix.loc[x.name, y.name]
    
    # 使用 matplotlib 的 colormap
    from matplotlib import cm
    cmap = sns.light_palette("blue", as_cmap=True)
    norm = plt.Normalize(0, 1)
    
    facecolor = cmap(norm(value))
    ax.set_facecolor(facecolor)
    
    # 根据背景亮度决定文字颜色
    text_color = "white" if value > 0.6 else "black"
    ax.text(0.5, 0.5, f'{value:.2f}',
            ha='center', va='center', fontsize=9,
            fontweight='bold', color=text_color, transform=ax.transAxes)

g.map_upper(plot_assoc_heatmap_absolute)

# 左下和对角线部分 (与之前版本相同)
def plot_lower_scatter(x, y, **kwargs):
    ax = plt.gca()
    is_numeric_x = is_numeric_dtype(x)
    is_numeric_y = is_numeric_dtype(y)
    if is_numeric_x and is_numeric_y:
        sns.scatterplot(x=x, y=y, ax=ax, s=10, alpha=0.6, **kwargs) # 减小点的大小
    else:
        sns.stripplot(x=x, y=y, ax=ax, jitter=0.25, s=3, alpha=0.7, **kwargs) # 减小点的大小
g.map_lower(plot_lower_scatter)

def plot_diag_dist(x, **kwargs):
    ax = plt.gca()
    if is_numeric_dtype(x):
        sns.histplot(x, kde=True, ax=ax, **kwargs)
    else:
        counts = x.value_counts()
        sns.barplot(x=counts.index, y=counts.values, ax=ax, order=counts.index, **kwargs)
        # 调整标签以适应更小的空间
        ax.tick_params(axis='x', labelrotation=45, labelsize=8)
g.map_diag(plot_diag_dist)

# --- 4. 性能与显示 ---
print("正在生成17x17矩阵图，这可能需要一些时间...")
plt.tight_layout()
plt.show()
print("绘图完成。")
```
